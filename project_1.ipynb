```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(dataset_1)
```

```python
X_scaled
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
for k in range(2, 8):
    km = KMeans(n_clusters=k, random_state=42)
    km.fit(X_scaled)
    inertia.append(km.inertia_)

plt.plot(range(2, 8), inertia)
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=42)
dataset_1['Risk_Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
cluster_profile = dataset_1.groupby('Risk_Cluster').mean()
cluster_profile
```

```python
dataset_1.groupby('Risk_Cluster').mean()
```

```python
risk_map = {
    0: "Low Risk",
    1: "Medium Risk",
    2: "High Risk",
    3: "No Credit History"
}

dataset_1['Risk_Label'] = dataset_1['Risk_Cluster'].map(risk_map)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2, random_state=42)
X_pca = pca.fit_transform(X_scaled)
```

```python
pca_df = pd.DataFrame(
    X_pca, 
    columns=['PC1', 'PC2']
)

pca_df['Risk_Cluster'] = dataset_1['Risk_Cluster']
pca_df['Risk_Label'] = dataset_1['Risk_Label']
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
for label in pca_df['Risk_Label'].unique():
    subset = pca_df[pca_df['Risk_Label'] == label]
    plt.scatter(subset['PC1'], subset['PC2'], label=label, alpha=0.6)

plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Risk Segmentation using PCA")
plt.legend()
plt.show()
```

```python
risk_score_map = {
    "Low Risk": 20,
    "Medium Risk": 50,
    "High Risk": 80,
    "No Credit History": 60
}

dataset_1['Risk_Score'] = dataset_1['Risk_Label'].map(risk_score_map)
```
